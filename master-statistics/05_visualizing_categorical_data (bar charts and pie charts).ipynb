```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Bar Chart

We are going to use a library called seaborn. It creates powerful visuals without any syntax headaches, and it is built off of the Matplotlib library.

We are going to use `.countplot()` method. It's not the only method that builds bar charts in the seaborn library. `barplot()` can also be used, and it is more flexible in that it can use any function to determine the height of the bars. Weâ€™ve chosen to use `.countplot()` because the syntax is simpler and does what we need!


```python
heroes_data = pd.read_csv("05_heroes_information.csv")
heroes_data_cleaned = heroes_data[heroes_data.Alignment != "-"]

sns.countplot(x = heroes_data_cleaned["Alignment"])
plt.show()
```

```python
# Order by ascending order

sns.countplot(x = heroes_data_cleaned["Alignment"], order=heroes_data_cleaned["Alignment"].value_counts(ascending=True).index)
plt.show()
```

```python
# Specify order (good for ordinal data)

sns.countplot(x = heroes_data_cleaned["Alignment"], order=["bad", "good", "neutral"])
plt.show()
```

# Pie Chart

```python
water_usage = pd.read_csv("05_water_usage.csv")
print(water_usage)

wedge_sizes = water_usage["prop"]
pie_labels = water_usage["water_source"]

plt.pie(wedge_sizes, labels=pie_labels)
plt.axis("equal")
plt.title("Distribution of House Water Usage")
plt.show()
```
