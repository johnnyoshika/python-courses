```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Got this data from Codeacademy course
students = pd.read_csv('students.csv')
students.head()
```

# Side by side boxplots

```python
# How do the scores differ between the 2 schools, GP and MS?

scores_GP = students.G3[students.school == 'GP']
scores_MS = students.G3[students.school == 'MS']

print('counts', len(scores_GP), len(scores_MS))

mean_GP = np.mean(scores_GP)
mean_MS = np.mean(scores_MS)
print('mean', mean_GP, mean_MS)
print('mean difference', mean_GP - mean_MS)

median_GP = np.median(scores_GP)
median_MS = np.median(scores_MS)
print('median', median_GP, median_MS)
print('median difference', median_GP - median_MS)
```

```python
sns.boxplot(data = students, x = 'school', y = 'G3')
plt.show()
```

```python
# How do the scores differ between the addresses (urban or rural)?

scores_urban = students.G3[students.address == 'U']
scores_rural = students.G3[students.address == 'R']

print('counts', len(scores_urban), len(scores_rural))

mean_urban = np.mean(scores_urban)
mean_rural = np.mean(scores_rural)
print('mean', mean_urban, mean_rural)
print('mean difference', mean_urban - mean_rural)

median_urban = np.median(scores_urban)
median_rural = np.median(scores_rural)
print('median', median_urban, median_rural)
print('median difference', median_urban - median_rural)
```

```python
sns.boxplot(data = students, x = 'address', y = 'G3')
plt.show()
```

# Overlapping Histogram

```python
# Important to use 'density' instead of 'frequency' when comparing histograms with different number of data points

plt.hist(scores_GP , color="blue", label="GP", density=True, alpha=0.5)
plt.hist(scores_MS , color="red", label="MS", density=True, alpha=0.5)
plt.legend()
plt.show()
```

```python
plt.hist(scores_urban , color="blue", label="urban", density=True, alpha=0.5)
plt.hist(scores_rural , color="red", label="rural", density=True, alpha=0.5)
plt.legend()
plt.show()
```
