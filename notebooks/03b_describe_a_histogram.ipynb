# Describe a Histogram

To describe the distribution of a dataset, you need to describe the following:

- Center
- Spread
- Skew
- Modality
- Outliers

Data source: https://data.cms.gov/Medicare-Inpatient/Inpatient-Prospective-Payment-System-IPPS-Provider/97k6-zzx3
Data only includes chest pain treatment (i.e. filtered by `cp_data[cp_data['DRG Definition'] == '313 - CHEST PAIN']`)

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
cp_data = pd.read_csv('chest_pain_treatment_costs.csv')
```

```python
cp_data.dtypes
```

```python
cp_data.head()
```

```python
ave_covered_charges = cp_data[' Average Covered Charges ']
```

## Center

```python
print('mean', ave_covered_charges.mean())
print('median', ave_covered_charges.median())
```

### Spread

```python
print('min', ave_covered_charges.min())
print('max', ave_covered_charges.max())
print('range', ave_covered_charges.max() - ave_covered_charges.min())
```

### Histogram

```python
plt.hist(ave_covered_charges, range=(0, 80000), bins=100, edgecolor='black')
plt.axvline(x=ave_covered_charges.mean(), label='Average', c='r')
plt.title('Cheat Pain Treatment')
plt.xlabel('Average Coverage charges')
plt.ylabel('Count')
plt.legend()
plt.show()
```

### Skew

Distribution is skewed right, b/c the tail goes far to the right and mean is greater than the median

### Modality

It's a unimodal distribution, as it only has one peak. If it had 2, it would have a bimodal distribution.

### Outlier

The larges cost is also an outlier
