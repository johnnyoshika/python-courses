# Histogram

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
exercise_ages = np.array([22, 27, 45, 62, 34, 52, 42,
                          22, 34, 26, 24, 65, 34, 25, 45, 23, 45, 33, 52, 55])

print('min', np.amin(exercise_ages))
print('max', np.amax(exercise_ages))
```

We can use `np.histogram()` to give us histogram info, but it's not as nice as a plot:

```python
np.histogram(exercise_ages, range=(20, 70), bins=5)
```

```python
plt.hist(exercise_ages, range=(20, 70), bins=5, edgecolor='black')

plt.title("Decade Frequency")
plt.xlabel("Ages")
plt.ylabel("Count")

plt.show()
```
